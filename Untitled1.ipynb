```python
import pandas as pd
from textblob import TextBlob


# 감정 분석 함수
def sentiment_textblob(text):
    try:
        x = TextBlob(text).sentiment.polarity
        if x < 0:
            return "Negative"
        elif x > 0:
            return "Positive"
        else:
            return "Neutral"
    except:
        return "Neutral"  # 예외 발생 시 중립으로 처리

# 감정 분포 계산 함수
def calculate_sentiment_distribution(df):
    if 'author' not in df.columns or 'text' not in df.columns:
        raise ValueError("The dataframe must contain 'author' and 'text' columns.")

    authors = df['author'].unique()
    sentiment_distribution = {}

    for author in authors:
        texts = df[df['author'] == author]['text']
        sentiments = texts.apply(sentiment_textblob)
        sentiment_counts = sentiments.value_counts()
        sentiment_distribution[author] = sentiment_counts.to_dict()

    return sentiment_distribution

# 감정 분포 계산
result = calculate_sentiment_distribution(train)

# 결과 출력
for author, distribution in result.items():
    print(f"Author {author}: {distribution}")
```

```python
# 특정 작가와 특정 감정에 해당하는 문장 필터링 함수
def filter_sentences_by_author_and_sentiment(df, author, sentiment_label):
    if 'author' not in df.columns or 'text' not in df.columns:
        raise ValueError("The dataframe must contain 'author' and 'text' columns.")

    # 해당 작가의 텍스트 추출
    author_texts = df[df['author'] == author].copy()  # 원본 데이터 수정 방지

    # 감정을 분석하여 새 컬럼에 저장
    author_texts.loc[:, 'sentiment'] = author_texts['text'].apply(sentiment_textblob)

    # 특정 감정에 해당하는 문장 필터링
    filtered_texts = author_texts[author_texts['sentiment'] == sentiment_label]
    return filtered_texts['text']

# 특정 작가(author=3)와 감정('Positive')에 해당하는 문장 보기
positive_sentences = filter_sentences_by_author_and_sentiment(train, author=3, sentiment_label='Positive')

# 출력: 상위 10개 문장
print("Positive Sentences for Author 3:")
print(positive_sentences.head(10).tolist())  # 상위 10개 문장을 리스트로 출력
```

```python
from transformers import Trainer, TrainingArguments, DataCollatorWithPadding
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# 평가 지표 정의
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = logits.argmax(axis=-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average="weighted")
    acc = accuracy_score(labels, predictions)
    return {"accuracy": acc, "f1": f1, "precision": precision, "recall": recall}

# 학습 설정
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,  # 안정적인 학습률
    per_device_train_batch_size=4,  # 작은 배치 크기
    gradient_accumulation_steps=4,  # 가상 배치 크기 효과
    num_train_epochs=3,  # 에포크 수 조정
    lr_scheduler_type="cosine",  # 학습률 스케줄러 추가
    save_total_limit=1,
    load_best_model_at_end=True
)

# 데이터 콜레이터
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics  # 성능 지표 함수 추가
)

# 모델 학습
trainer.train()
```
